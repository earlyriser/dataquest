```python
# 1. MOVIE REVIEWS
import pandas as pd

movies = pd.read_csv("fandango_score_comparison.csv")
movies
```

```python
# 2. Histograms
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(movies['Metacritic_norm_round'])
plt.show()
plt.hist(movies['Fandango_Stars'])
plt.show()
```

It seems **Fandango** ratings are just occuring from 3 to 5, having mostly neutral and positive ratings.

On the ohter hand, **Metacritic** has negative and positive ratings.

```python
# 3. Mean, median, and standard deviation
f_mean = movies['Fandango_Stars'].mean()
m_mean = movies['Metacritic_norm_round'].mean()

f_median = movies['Fandango_Stars'].median()
m_median = movies['Metacritic_norm_round'].median()

f_std = movies['Fandango_Stars'].std()
m_std = movies['Metacritic_norm_round'].std()

print(f_mean)
print(m_mean)
print(f_median)
print(m_median)
print(f_std)
print(m_std)
```

|                    | Fandango | Metacritic |
|--------------------|----------|------------|
| Mean               |   4.08   |    2.97    |
| Median             |  4       |       3    |
| Standard Deviation |   0.54   |   0.99     |
|                    |          |            |

Metacritic ratings is an aggregations of ratings from multiple websites, and they give more weight to critics with good reputation

Fandango is the aggregation of raitings of people watching the movies, not really critics. Then is normal that a person paid to see a movie, already likes the actors, directors or the style.

So critics are not predetermined to love a movie, but moviegoers have already voted with their dollars about the hunch that a movie is going to be good.

The **standard deviation** of Fandango is lower than Metacritic one, showing that the results are biased to to right.

```python
# 4. Scatter plots
# Checking for outliers
plt.figure(figsize=(6,6))
plt.scatter(movies['Fandango_Stars'], movies['Metacritic_norm_round'])
plt.xlim(0,5.5)
plt.ylim(0,5.5)
plt.show()

```

```python
# Finding the movies with the biggest discrepancy between Fandango & Metacritic
movies['fm_diff']=movies['Metacritic_norm_round']-movies['Fandango_Stars']
movies['fm_diff'] = abs(movies['fm_diff'])
movies['fm_diff']
movies.sort_values(by=['fm_diff'], ascending=[False])
movies.head()
```

```python
# 5. Correlations
# R value
from scipy.stats import pearsonr

xv = movies['Metacritic_norm_round']
yv = movies['Fandango_Stars']

r_value, p= pearsonr(xv,yv)
r_value
```

r-value between Fandango and Metacritic is very low: **0.178**

```python
# Linear regression
from scipy.stats import linregress

slope, intercept, r_value, p_value, std_err = linregress(xv, yv)

print(slope, intercept, r_value, p_value)
```

```python
# A movie with a 3 in Metacritic gets what on Fandango
def compute_predicted_y(x):
    return x * slope + intercept

print(compute_predicted_y(1))
print(compute_predicted_y(2))
print(compute_predicted_y(3))
print(compute_predicted_y(4))
print(compute_predicted_y(5))
```
